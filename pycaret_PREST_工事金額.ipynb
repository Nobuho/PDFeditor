```
from pycaret import datasets
import pandas as pd
import numpy as np

df = pd.read_csv("prest_kojikingaku.csv")
dataset = df
dataset.info()
```

```
import pandas_profiling
dataset.profile_report()
```

```
data = dataset.sample(frac=0.9, random_state=786).reset_index(drop=True)
data_unseen = dataset.drop(data.index).reset_index(drop=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```
from pycaret.regression import *
# 特徴量等の調節はここで行う
exp_reg101 = setup(
    data=data,
    target='kanseikouzigenka',
    categorical_features=['gaiheki',"kouzou1","sekkeikubun","tokuirekubun"],
    numeric_features=[
        "kaisuB",
        "kaisuF"],
    ignore_features=["tubotanka","kouzisyubetu1"],
    high_cardinality_features=['kankatumeisyou'])
```

```
compare_models()
```

```
import time

start = time.time()

# 選んだモデル
select = "huber"
# チューニング
tuned_my_model = tune_model(select)
# ファイナライズ
final_my_model = finalize_model(tuned_my_model)
unseen_predictions = predict_model(final_my_model, data=data_unseen)
save_model(tuned_my_model, 'prest_kojigenka_model_200502')

elapsed_time = time.time() - start
print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
```
